```python
import pandas as pd
data_crime = pd.read_csv('F:\Crime rate prediction\Working in colab\crime_state_number.csv')
data_crime
```

```python
data_state_number = data_crime[['STATE','MURDER','YEAR']]
data_state_number
```

```python
data_state = data_state_number[data_state_number['STATE']==10]
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np


plt.scatter(x =data_state.YEAR , y =data_state.MURDER , marker='o');
plt.xlabel('YEAR')
plt.ylabel('MURDER')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x =data_state.YEAR
y =data_state.MURDER
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
