```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

from subprocess import check_output

crimes1 = pd.read_csv('F:\Crime rate prediction\Working in colab/01_District_wise_crimes_committed_IPC_2001_2012.csv')
crimes2 = pd.read_csv('F:\Crime rate prediction\Working in colab/01_District_wise_crimes_committed_IPC_2013.csv')

crimes = pd.concat([crimes1,crimes2],  ignore_index=False, axis=0)

crimes.rename(columns={'STATE/UT':'STATE'}, inplace=True)

del crimes1
del crimes2


print('Dataset is ready....')
crimes.shape


states = crimes.STATE.unique()


for i in range(0, len(states)):
    states[i] = states[i].lower()
    
for j in range(0, len(states)):
    if states[j] == 'a & n islands':
       states[j] = 'a&n islands'
    
    if  states[j] == 'd & n haveli':
        states[j] = 'd&n haveli'
      
print(states)


states = np.unique(states).tolist()

print(states)

crimes['STATE'] = crimes['STATE'].str.lower()
crimes.head(3)


crimes_total = crimes[crimes['DISTRICT'] == 'TOTAL']


crimes_total.drop('DISTRICT', axis=1, inplace=True)

crimes_total = crimes_total[crimes_total['YEAR'] != 2013]
crimes_total
```

```python
crimes_totals = crimes_total[['STATE','YEAR','MURDER','RAPE','ARSON','THEFT','KIDNAPPING & ABDUCTION']]
crimes_totals
```
