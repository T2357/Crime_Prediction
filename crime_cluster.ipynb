```python
import pandas as pd
data_crime = pd.read_csv('F:\Crime rate prediction\Working in colab\crime_state_name.csv')
data_crime
```

```python
data1 = data_crime.loc[:,['STATE', 'YEAR','MURDER','RAPE','THEFT','RIOTS','DOWRY DEATHS','ARSON']]
data1.head(5)
```

```python
data2 = data1.loc[data1['YEAR'] ==2012]
data2.head(5)
```

```python
data3 = data2
data3
```

```python

STATE = {'andhra pradesh':1, 'arunachal pradesh':2, 'assam':3, 'bihar':4, 'chhattisgarh':5, 'goa':6, 'gujarat':7, 'haryana':8,'himachal pradesh':9,'jammu & kashmir':10,'jharkhand':11,'karnataka':12, 'kerala':13, 'madhya pradesh':14, 'maharashtra':15, 'manipur':16, 'meghalaya':17,
'mizoram':18, 'nagaland':19, 'odisha':20, 'punjab':21, 'rajasthan':22, 'sikkim':23, 'tamil nadu':24,
 'tripura':25, 'uttar pradesh':26, 'uttarakhand':27, 'west bengal':28, 'a & n islands':29,
'chandigarh':30, 'd & n haveli':31, 'daman & diu':32, 'delhi ut':33, 'lakshadweep':34,'puducherry':35
} 
data3.STATE = [STATE[item] for item in data3.STATE]
data3
```

```python
data3['STATE'] = data3['STATE'].astype('category')
data3.dtypes
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np


plt.scatter(x =data3.STATE , y =data3.MURDER , marker='o');
plt.xlabel('STATE')
plt.ylabel('MURDER')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x =data3.STATE
y =data3.MURDER
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = data3.iloc[:, [2,0]].values
x
```

```python
kmeans3 = KMeans(n_clusters=3)
y_kmeans3 = kmeans3.fit_predict(x)
print(y_kmeans3)
```

```python
kmeans3.cluster_centers_
```

```python
Error =[]
for i in range(1, 7):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)
import matplotlib.pyplot as plt
plt.plot(range(1, 7), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

```python
plt.scatter(x[:,1],x[:,0],c=y_kmeans3,cmap='rainbow')
```
